```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Load first 50,000 rows

df = pd.read_csv("Reviews.csv", nrows=50000)

print("Dataset Loaded Successfully")
print("Shape:", df.shape)
```

```python
# Find reviews containing the word "light"

light_reviews = df[df['Text'].str.contains(r'\blight\b', case=False, na=False)]

print("Total reviews containing 'light':", len(light_reviews))

for i, review in enumerate(light_reviews['Text'].head(20)):
    print(f"\nReview {i+1}:")
    print(review)
```

```python
sentence_1 = "This olive oil is wonderful and very light on the stomach."

sentence_2 = "The packaging was broken and let too much light inside the box."

sentence_3 = "I prefer this brand because it is a light and healthy snack."

documents = [sentence_1, sentence_2, sentence_3]

print("Sentence 1:", sentence_1)
print("Sentence 2:", sentence_2)
print("Sentence 3:", sentence_3)
```

```python
# Vectorization

vectorizer = CountVectorizer()

document_term_matrix = vectorizer.fit_transform(documents)

dense_matrix = document_term_matrix.toarray()

print("Vocabulary:")
print(vectorizer.get_feature_names_out())

print("\nDense Matrix:")
print(dense_matrix)
```

```python
# Extract vectors

vec_1 = dense_matrix[0]
vec_2 = dense_matrix[1]
vec_3 = dense_matrix[2]

print("Vector 1:", vec_1)
print("Vector 2:", vec_2)
print("Vector 3:", vec_3)
```

```python
# Dot Product Similarities

similarity_1_2 = np.dot(vec_1, vec_2)

similarity_1_3 = np.dot(vec_1, vec_3)

print("Sentence 1 vs Sentence 2 Similarity:", similarity_1_2)

print("Sentence 1 vs Sentence 3 Similarity:", similarity_1_3)
```
